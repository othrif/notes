### Line plot

```python
x = [1, 2, 3, 4,10]
y = [10, 15, 17, 20,30]
plt.plot(x,y)

plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('My Title')
plt.yticks([0,5,10,15,20,25,30,40],
          ['zero', 'five', 'ten', 'fifteen', 'twenty', 'twenty five', 'thirty', 'fourty'])

plt.show()
```

### Scatter plot

```python
plt.clf()
plt.scatter(x,y)
plt.show()
```

#### Change size of bullets

```python
plt.clf()
size = [100,30,2000,50,1000]
plt.scatter(x,y,s=size)
```

#### Add colours

```python
plt.clf()
col = ['yellow', 'orange', 'red', 'blue', 'violet']
plt.scatter(x,y,s=size,c=col)
```

#### make transparent

```python
plt.clf()
col = ['yellow', 'orange', 'red', 'blue', 'violet']
plt.scatter(x,y,s=size,c=col, alpha=0.5)
```

#### Add text and grid

```python
plt.clf()
col = ['yellow', 'orange', 'red', 'blue', 'violet']
plt.scatter(x,y,s=size,c=col, alpha=0.5)

# Additional customizations
plt.text(2, 10, 'India')
plt.text(6, 25, 'China')

# Add grid() call
plt.grid(True)

plt.show()
```
